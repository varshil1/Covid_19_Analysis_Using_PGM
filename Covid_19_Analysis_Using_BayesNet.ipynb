**Now we will convert the values in three categories i.e. 0,1 and 2**

We convert gender column name as here 

0: female , 1 : male

```
gender = {2: 1, 1:0} #1 as female(0) , 2 as male(1)
df['sex']=df['sex'].replace([2,1],[1,0])
df.head()
```

```
type = {2: 1, 1:0} # 1 for not hospitalized(0) and 2 for hosptalized(1).
df['patient_type']=df['patient_type'].replace([2,1],[1,0])
df.head()
```

We convert date_died column name as here

```
 #skip entry date and date symtoms
 # For date -died 9999-99-99 is patient recovered(0) and for others patient died(1)
df['date_died']=df['date_died'].apply(lambda x: 0 if x=='9999-99-99' else 1)
df.head()
```

We convert intubed column name as here

0: not used ventilator , 1 : used ventilator ,2: not specified


```
#intubed : Intubation is a procedure that's used when you can't breathe on your own. Your doctor puts a tube down your throat and into your windpipe to make it easier to get air into and out of your lungs. A machine called a ventilator pumps in air with extra oxygen. Then it helps you breathe out air that’s full of carbon dioxide (CO2). "1" denotes that the patient used ventilator and "2" denotes that the patient did not, "97" "98" "99" means not specified.
#0 not used ventilator,1 used ventilator,2 not speciefied
df['intubed']=df['intubed'].apply(lambda x: 0 if x==2 else(1 if x==1 else 2))
df.head()
```

**Making Bayesian model**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import networkx as nx
# model = BayesianModel([('patient_type', 'renal_chronic'),('patient_type', 'inmsupr'),('patient_type', 'other_disease'),('patient_type', 'cardiovascular'), ('patient_type', 'hypertension'),('patient_type', 'asthma'), ('patient_type', 'copd'),('patient_type', 'icu'),
#                        ('icu', 'diabetes'),('icu', 'sex'),('icu', 'date_died'),('icu', 'pneumonia'),
#                        ('diabetes', 'hypertension'),('diabetes', 'renal_chronic'),('diabetes', 'copd'),('diabetes', 'date_died'),('diabetes', 'pneumonia'),
#                        ('hypertension', 'renal_chronic'),('hypertension', 'cardiovascular'),('hypertension', 'obesity'),('hypertension', 'date_died'),('hypertension', 'copd'),
#                        ('renal_chronic', 'inmsupr'),('renal_chronic', 'cardiovascular'),
#                        ('cardiovascular', 'other_disease'),('cardiovascular', 'obesity'),('cardiovascular', 'copd'),
#                        ('obesity', 'asthma'),
#                        ('asthma', 'inmsupr'),('asthma', 'sex'),
#                        ('inmsupr', 'other_disease'),
#                        ('other_disease', 'sex'),
#                        ('sex', 'date_died'),
#                        ('date_died', 'pneumonia'),])

model=BayesianModel([('diabetes','obesity'),('diabetes','date_died'),('diabetes','pneumonia'),('diabetes','age'),('diabetes','patient_type'),('diabetes','renal_chronic'),('diabetes','hypertension'),
                     ('hypertension','age'),('hypertension','copd'),('hypertension','patient_type'),('hypertension','cardiovascular'),('hypertension','renal_chronic'),
                     ('age','asthma'),('age','copd'),('age','pneumonia'),('age','obesity'),('age','inmsupr'),('age','cardiovascular'),('age','patient_type'),('age','renal_chronic'),
                     ('patient_type','copd'),('patient_type','icu'),('patient_type','other_disease'),('patient_type','cardiovascular'),('patient_type','inmsupr'),('patient_type','copd'),('patient_type','renal_chronic'),
                     ('renal_chronic','obesity'),('renal_chronic','cardiovascular'),('renal_chronic','inmsupr'),
                     ('obesity','asthma'),('obesity','icu'),
                     ('cardiovascular','copd'),('cardiovascular','other_disease'),
                     ('copd','asthma'),
                     ('icu','pneumonia'),('icu','date_died'),('icu','sex'),
                     ('asthma','sex'),('asthma','inmsupr'),
                     ('inmsupr','other_disease'),
                     ('pneumonia','date_died'),
                     ('date_died','sex'),
                     ('other_disease','sex')
                     ])
model.fit(df, estimator=MaximumLikelihoodEstimator)
```

# Deriving Inferences from Bayesian model

```
#Create Inference Object of the model and run VE algorithm on it
inference=VariableElimination(model)

#Query1 : Given that the patient's age is greater than 42(mean age),is in icu,has diabetes, but don't have hyper tension and cardiovascular disease, will the patient die due to corona?
query1=inference.map_query(variables=['date_died'], evidence={'age':1,'icu':1,'diabetes':1,'hypertension':0,'cardiovascular':0})
print(query1)
```

```

#Query2 : Given that the patient is diabetic,is admitted to icu and is examined with pneumonia and cardiovascular disease, will the patient die due to corona?
query2=inference.map_query(variables=['date_died'], evidence={'diabetes':1,'icu':1,'pneumonia':1,'cardiovascular':1}) 
print(query2)

```

```
#Query3 : Given that the patient is diabetic,is admitted to icu ,doesn't suffer from hypertension,examined with obesity and has age grater than 42(mean), will the patient die due to corona?
query3=inference.map_query(variables=['date_died'], evidence={'diabetes':1,'icu':1,'hypertension':0,'obesity':1,'age':0})
print(query3)

```

```
#Query4 : Given that the patient is diabetic,but not examined with renal chronic and copd disease, will the patient die due to corona?
query4=inference.map_query(variables=['date_died'], evidence={'diabetes':1,'renal_chronic':0,'copd':0})
print(query4)
```

```
#Query5 : Given that the patient is female,has age > 42,is hospitalized in icu,is diabetic,suffer from hypertension and cardivascular disease but do not have copd,asthma,obesity,renal chronic diseases, will the patient die due to corona?

query5=inference.map_query(variables=['date_died'], evidence={'sex':0,'age':1,'patient_type':1,'pneumonia':0,'diabetes':1,'copd':0,'asthma':0,'inmsupr':0,'hypertension':1,'other_disease':0,'cardiovascular':1,'obesity':0,'renal_chronic':0,'icu':1})
print(query5)
```

```
#Query6 : Given that the patient do no suffer from obesity,asthma,pneumonia, but suffer from hypertension,has age > 42,and is hospitalized needs to admit to ICU, to prevent death?
query6=inference.map_query(variables=['icu'], evidence={'obesity':0,'date_died':0,'age':1,'asthma':0,'pneumonia':0,'patient_type':1,'hypertension':1})
print(query6)
```

```
#Query7 : Given that the patient is dead due to corona, what will be the age and sex of person?
query7=inference.map_query(variables=['age','sex'], evidence={'date_died':1})
print(query7)
```

```
#Query8 : Given that the patient is female and is dead due to corona, what will be the age person and would he/she be suffered by pnuemonia,diabetes,cardiovascular,obesity,asthma and hypertension?
query8=inference.map_query(variables=['age','pneumonia','diabetes','cardiovascular','obesity','asthma','hypertension'], evidence={'date_died':1,'sex':1})
print(query8)
```

```
#Query9 : Given that the patient is diabetic,admitted to ICU,suffers from pneumonia,hypertension,is a male, and cardiovascular disease is not specified ,will the patient die due to corona ?
query9=inference.map_query(variables=['date_died'], evidence={'diabetes':1,'icu':1,'pneumonia':1,'sex':1,'cardiovascular':2,'hypertension':1}) 
print(query9)
```

```
#Query10 : Given that the patient is diabetic , admitted to icu,is alive, male patient with hypertension and copd but no obesity, will the patient suffers from pneumonia?
query10=inference.map_query(variables=['pneumonia'], evidence={'diabetes':1,'icu':1,'date_died':0,'sex':1,'hypertension':1,'obesity':0,'copd':1}) 
print(query10)
```
